```python
imdb_top250 = pd.read_html("https://www.imdb.com/chart/top?ref_=nv_wl_img_3")[0]
imdb_top250.head()
```

```python
del imdb_top250['Unnamed: 0']
del imdb_top250['Unnamed: 4']
del imdb_top250['Your Rating']
```

```python
imdb_top250.head()
```

```python
imdb_top250['Year'] = imdb_top250['Rank & Title'].str[-5:-1] # take the Year out of the Rank&Title Column
imdb_top250['Rank & Title'] = imdb_top250['Rank & Title'].str[0:-6]
imdb_top250.head()
```

```python
imdb_top250['Ranking'] = imdb_top250['Rank & Title'].str.extract('(\d{1,4}).\ ')
imdb_top250['Title'] = imdb_top250['Rank & Title'].str.extract('\d{1,4}.\ (.*)')
imdb_top250.drop(['Rank & Title'], axis=1, inplace=True)
imdb_top250.head()
```

```python
imdb_2010 = imdb_top250[imdb_top250['Year'] == '2010']
imdb_2010_sorted = imdb_2010.sort_values('IMDb Rating', ascending=False)
imdb_2010_sorted
```

```python
# aggregate the highest average rankings for each year
movies_by_year = imdb_top250.groupby('Year').mean()
# Highest ranked year
movies_by_year.loc[movies_by_year['IMDb Rating'].idxmax()]
```
